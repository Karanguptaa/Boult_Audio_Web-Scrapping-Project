```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.boultaudio.com/products/z40?variant=40466476171333"
```

```python
r = requests.get(url)
htmlcontent = r.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
#print(soup)
soup.prettify #Prettify the codes of the HTMl page
```

```python
title = soup.title
print(type(soup)) # BeautifulSoup
print(type(title)) # Tag
print(type(title.string)) # Navigablestring
```

```python
title=soup.title
title
```

```python
Price = soup.find_all("div",class_="WI_pT_price")
Price
```

```python
Product_Price = []
for i in range(0,len(Price)):
    Product_Price.append(Price[i].get_text())
Product_Price
```

```python
Price = soup.find("div",class_="WI_floatingPrice").text
Price.replace('Sale price','').strip()
```

```python
colour = soup.find("div",class_="WI_pT_option").text
colour.strip().replace('colors',' ')
```

```python
feature = soup.find("div",class_="WI_productDrop_main").text
feature.strip().replace('\n\n\n\n\n\n\n\n\n\n\n','   ')
```

Access Review

```python
Comment = soup.find_all("div",class_="jdgm-rev__body")
Comment
```

```python
Review_Comment = []
for i in range(0,len(Comment)):
    Review_Comment.append(Comment[i].get_text())
Review_Comment
```

Access Customers Names

```python
customers = soup.find_all("span",class_="jdgm-rev__author-wrapper")
customers
```

```python
cust_name = []
for i in range(0,len(customers)):
    cust_name.append(customers[i].get_text())
cust_name
```

Access Ratings

```python
rating_one = soup.find("span",class_="jdgm-rev__rating")['data-score']
rating_one
```

```python
ratings = [span['aria-label']for span in soup.find_all("span",class_="jdgm-rev__rating")]
ratings
```

Arranging all Data using Pandas Libraray

```python
import pandas as pd
df = pd.DataFrame({"Review_Comments": Review_Comment,
                   "Customers_Name": cust_name,
                   "Review_Stars": ratings})
print(df)
```

```python
df.info()
```

Export Data to CSV

```python
df.to_csv("Rynox_Gear_reviews.csv",index = False)
```
